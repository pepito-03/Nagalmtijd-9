```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#clear old figures
#plt.close("all")

#load CSV
df = pd.read_csv("kerk1.csv", header=None)

#create figure
fig, ax = plt.subplots(figsize=(12, 8))

#plot heatmap
im = ax.imshow(df.values, cmap="inferno", aspect="auto",
               origin="lower", vmin=0, vmax=5)
#auto x/y labels
ax.set_xticks(range(df.shape[1]))
ax.set_yticks(range(df.shape[0]))
ax.set_xticklabels([f"x{i+1}" for i in range(df.shape[1])])
ax.set_yticklabels([f"y{i+1}" for i in range(df.shape[0])])

#overlay numbers
for i in range(df.shape[0]):
    for j in range(df.shape[1]):
        # white text for dark cells, black for light cells
        color = "white" if df.iloc[i, j] > df.values.max() / 2 else "black"
        ax.text(j, i, round(df.iloc[i, j], 2), ha="center", va="center", color=color)

#attach colorbar
fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

#show
plt.imshow(df, origin='lower', cmap='inferno')
plt.colorbar(label="Geluidsintensiteit (dB)")
plt.xlabel("x-positie")
plt.ylabel("y-positie")
plt.title("Geluidsintensiteit heatmap")


plt.show()
```
